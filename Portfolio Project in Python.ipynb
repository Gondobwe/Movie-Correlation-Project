```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of the plots we will create

#Read in the data

df = pd.read_csv(r'C:\Users\Hp\Desktop\Portfolio Project 1\archive\movies.csv')

```

```python
#Looking at the data

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
df = df.dropna()
```

```python
df.dtypes
```

```python
#Specify the column and change the data type

df = df.fillna(0)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['company'] = df['company'].astype('str')

```

```python
df
```

```python
#Create correct Year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
df
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# These are the assumptions being investigated
# Budget high correlation
# Company High correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earning')

plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# Looking at the correlation in detail

df.corr()  #By default this shows Pearson correlation, there are also Kendall and Spearman
```

```python
df.corr(method = 'kendall')
```

```python
df.corr(method = 'spearman')
```

```python
# There is high correlation between budget and gross
```

```python
#Visualising the correlation matrix

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#Unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
